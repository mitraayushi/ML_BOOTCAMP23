```python
#REQUIRED LIBRARIES
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#READING THE DATAFRAME
df = pd.read_csv('/content/Salary_dataset (1).csv')
```

# **PRE-PROCESSING THE GIVEN DATAFRAME**

```python
#TO DISPLAY TOP 5 ROWS
df.head(5)
```

```python
#TO DISPLAY BOTTOM 5 ROWS
df.tail(5)
```

# LINEAR REGRESSION



```python
#DROPPING NULL VALUES(IF ANY)
df = df.dropna()
```

```python
#DROPPING THE EXTRA COLUMN PRESENT IN THE DATAFRAME
df=df.drop('Unnamed: 0', axis = 1)
```

```python
#DISPLAYING THE PROCESSED DATASET
df
```

PLOTTING OF GRAPH

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.xlabel('YEARS OF EXPERIENCE')
plt.ylabel('SALARY')
```

```python
#SPLITTING THE DATAFRAME AS X(YEARS OF DATAFRAME) AND Y(SALARY)
X = df.iloc[:,0:1]
Y = df.iloc[:,1]
```

```python
#SHOWING THE VALUES IN X
X
```

```python
#SHOWING THE VALUES IN Y
Y
```

```python
#IMPORT OF LIBRARY
from sklearn.model_selection import train_test_split
#DIVIDING THE DATAFRAME INTO TRAIN AND TEST
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
#IMPORT OF LIBRARY
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
#TRAINING THE LR MODEL
lr.fit(X_train,Y_train)
```

```python
X_test
```

```python
Y_test
```

```python
#PREDICTION OF VALUES
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
#PLOTTING THE GRAPH TO SHOW REGRESSION LINE AND DATA POINTS AND EXAMINE THE MODEL'S FIT
plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('YEARS OF EXPERIENCE')
plt.ylabel('SALARY')
```

```python
#TO CHECK ACCURACY
from sklearn.metrics import r2_score
```

```python
#EXAMININING THE MODEL'S PERFORMANCE AND CHECKING IT'S ACCURACY
Y_true = Y_test
Y_pred = lr.predict(X_test)
r2_score(Y_true,Y_pred)
```

```python
#THE FORMULA y=mx+c IS UDED IN THIS MODEL WHERE m IS THE SLOPE AND c IS THE INTERCEPT
m = lr.coef_
```

```python
c = lr.intercept_
```

```python
#EXAMPLE---- y(PREDICTED SALARY) for x=8(YEARS OF EXPERIENCE)
y = m*8 + c
y
```
